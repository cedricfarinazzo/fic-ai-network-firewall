```python
from tqdm import tqdm
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
## Loading data
```

```python
X_train, X_test, Y_train, Y_test = load_dataset(test_size=0.1)
```

```python
## Model
```

```python
rfc = RandomForestClassifier(warm_start = True, n_estimators = 1500)
```

```python
## Train
```

```python
chunk_size = 4096
cycle = 10
total_x_train = len(X_train)
total_chunks = total_x_train / chunk_size

with tqdm(total=total_chunks * cycle) as pbar:
    for _ in range(cycle):
        for chunk in range(0, total_x_train, chunk_size):
            arr_X = load_chunk(X_train[chunk:chunk + chunk_size])
            arr_Y = Y_train[chunk:chunk + chunk_size]
            
            rfc.fit(arr_X, arr_Y)
            rfc.n_estimators += 1

            pbar.update(1)
```

```python
## Metrics
```

```python
chunk_size=1024
total_x_test = len(X_test)
total_chunks = total_x_test / chunk_size

Y_pred = np.array([])
with tqdm(total=total_chunks) as pbar:
    for chunk in range(0, total_x_test, chunk_size):
        arr_X = load_chunk(X_test[chunk:chunk + chunk_size])
       
        Y_pred = np.concatenate((Y_pred, rfc.predict(arr_X)), axis=0)
        pbar.update(1)

count_misclassified = (Y_test != Y_pred).sum()
```

```python
total_y = len(Y_test)
print(f"Misclassified samples: {count_misclassified}/{total_y}")
accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy: {:.2f}'.format(accuracy))
```
